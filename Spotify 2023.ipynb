# `                                   DATA ANALYSIS                                                                  `

```python
plt.figure(figsize=(20,12))
plt.subplot(2,2,1)
sns.countplot(y=df.artist_name,palette='flare',order=df.artist_name.value_counts().index[:11])
plt.xticks(rotation=90)
plt.title('Top 10 artists',fontsize=25,fontname="Gabriola")

plt.subplot(2,2,2)
sns.countplot(y=df.track_name,palette='flare',order=df.track_name.value_counts().index[:11])
plt.title('Top 10 Tracks',fontsize=25,fontname="Gabriola")
plt.xticks(rotation=90)

plt.subplot(2,2,3)
popular_artist = df.groupby('artist_name')['popularity'].sum().reset_index().sort_values(by='popularity',ascending=False)[:11]
sns.barplot(x=popular_artist.popularity,y=popular_artist.artist_name,palette='flare')
plt.title('Top 10 Popular Artists',fontsize=25,fontname="Gabriola")
plt.xticks(rotation=90)

plt.subplot(2,2,4)
popular_tracks = df.groupby('track_name')['popularity'].sum().reset_index().sort_values(by='popularity',ascending=False)[:11]
sns.barplot(x=popular_tracks.popularity,y=popular_tracks.track_name,palette='flare')
plt.title('Top 10 Popular Tracks',fontsize=25,fontname="Gabriola")
plt.xticks(rotation=90)
sns.despine()
plt.show()
```

```python
# years and different artists
plt.figure(figsize=(20,5))
years_count2 = df.groupby('year')['track_name'].count().reset_index()
sns.barplot(x=years_count2.year,y=years_count2.track_name,palette='flare')
plt.title('Number of tracks every year',fontsize=25,fontname="Gabriola")
plt.show()
```

```python
# top genre for danceability
genre_dance = df.groupby('genre')['danceability'].sum().reset_index().sort_values(by='danceability',ascending=False)[:6]
# top genre for energy
genre_energy = df.groupby('genre')['energy'].sum().reset_index().sort_values(by='energy',ascending=False)[:6]
#  top genre for speechiness
genre_speech = df.groupby('genre')['speechiness'].sum().reset_index().sort_values(by='speechiness',ascending=False)[:6]
# top genre for liveliness
genre_live = df.groupby('genre')['liveness'].sum().reset_index().sort_values(by='liveness',ascending=False)[:6]
#  top genre for valence
genre_valence = df.groupby('genre')['valence'].sum().reset_index().sort_values(by='valence',ascending=False)[:6]
# top genre for instrumentalness
genre_instruments = df.groupby('genre')['instrumentalness'].sum().reset_index().sort_values(by='instrumentalness',ascending=False)[:6]
```

```python
plt.figure(figsize=(20,12))
plt.subplot(2,3,1)
sns.barplot(x=genre_dance.genre,y=genre_dance.danceability,palette='flare')
plt.title('Top 5 genres associated with a high degree of danceability',fontsize=20,fontname="Gabriola")
plt.subplot(2,3,2)
sns.barplot(x=genre_energy.genre,y=genre_energy.energy,palette='flare')
plt.title('Top 5 genres associated with a high degree of energy',fontsize=20,fontname="Gabriola")
plt.subplot(2,3,3)
sns.barplot(x=genre_speech.genre,y=genre_speech.speechiness,palette='flare')
plt.title('Top 5 genres associated with a high degree of speechiness',fontsize=20,fontname="Gabriola")
plt.subplot(2,3,4)
sns.barplot(x=genre_live.genre,y=genre_live.liveness,palette='flare')
plt.title('Top 5 genres associated with a high degree of liveness',fontsize=20,fontname="Gabriola")
plt.subplot(2,3,5)
sns.barplot(x=genre_valence.genre,y=genre_valence.valence,palette='flare')
plt.title('Top 5 genres associated with a high degree of valence',fontsize=20,fontname="Gabriola")
plt.subplot(2,3,6)
sns.barplot(x=genre_instruments.genre,y=genre_instruments.instrumentalness,palette='flare')
plt.title('Top 5 genres associated with a high degree of instrumentalness',fontsize=20,fontname="Gabriola")
sns.despine()
plt.show()
```

```python
df['duration_min'] = df['duration_ms']/60000
genre_duration = df.groupby('genre')['duration_min'].sum().reset_index().sort_values(by='duration_min',ascending=False)
genre_duration_mean = df.groupby('genre')['duration_min'].mean().reset_index().sort_values(by='duration_min',ascending=False)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
sns.barplot(y=genre_duration.genre[:11],x=genre_duration.duration_min[:11],palette='flare')
plt.title('Top 10 genres based on their duration in minutes',fontname="Gabriola",fontsize=20)
plt.xticks(rotation=90)
plt.subplot(1,2,2)
sns.barplot(y=genre_duration.genre[77:83],x=genre_duration.duration_min[77:83],palette='flare')
plt.title('Bottom 5 genres based on their duration in minutes',fontname="Gabriola",fontsize=20)

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
sns.barplot(y=genre_duration_mean.genre[:11],x=genre_duration_mean.duration_min[:11],palette='flare')
plt.title('Top 10 genres based on their mean duration in min',fontname="Gabriola",fontsize=20)
plt.xticks(rotation=90)
plt.subplot(1,2,2)
sns.barplot(y=genre_duration_mean.genre[77:83],x=genre_duration_mean.duration_min[77:83],palette='flare')
plt.title('Bottom 5 genres based on their mean duration in min',fontname="Gabriola",fontsize=20)

plt.xticks(rotation=90)
plt.show()
```

```python
per_year = df.groupby('year')['popularity','danceability','energy','loudness','speechiness',
                                        'acousticness','instrumentalness','liveness','valence','tempo'].mean().reset_index()

# Calculate the percentage increase in popularity from one year to the next
per_year['popularity_change'] = (per_year['popularity'].pct_change() * 100).round(2)
per_year['danceability_change'] = (per_year['danceability'].pct_change() * 100).round(2)
per_year['energy_change'] = (per_year['energy'].pct_change() * 100).round(2)
per_year['loudness_change'] = (per_year['loudness'].pct_change() * 100).round(2)
per_year['speechiness_change'] = (per_year['speechiness'].pct_change() * 100).round(2)
per_year['acousticness_change'] = (per_year['acousticness'].pct_change() * 100).round(2)
per_year['instrumentalness_change'] = (per_year['instrumentalness'].pct_change() * 100).round(2)
per_year['liveness_change'] = (per_year['liveness'].pct_change() * 100).round(2)
per_year['valence_change'] = (per_year['valence'].pct_change() * 100).round(2)
per_year['tempo_change'] = (per_year['tempo'].pct_change() * 100).round(2)
per_year.head()
```

```python
plt.figure(figsize=(18,30))
plt.subplot(5,2,1)
sns.barplot(x=per_year['year'],y=per_year['popularity_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,2)
sns.barplot(x=per_year['year'],y=per_year['danceability_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,3)
sns.barplot(x=per_year['year'],y=per_year['energy_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,4)
sns.barplot(x=per_year['year'],y=per_year['loudness_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,5)
sns.barplot(x=per_year['year'],y=per_year['speechiness_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,6)
sns.barplot(x=per_year['year'],y=per_year['acousticness_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,7)
sns.barplot(x=per_year['year'],y=per_year['instrumentalness_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,8)
sns.barplot(x=per_year['year'],y=per_year['liveness_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,9)
sns.barplot(x=per_year['year'],y=per_year['valence_change'],palette='flare')
plt.xticks(rotation=45)
plt.subplot(5,2,10)
sns.barplot(x=per_year['year'],y=per_year['tempo_change'],palette='flare')
plt.xticks(rotation=45)
sns.despine()
plt.show()
```
